```python
zs = h.iloc[:,3]
print('La media del campo magnético es: ' + str(stats.mean(zs)))
print('La desviación del campo magnético es: ' + str(stats.pstdev(zs)))
media = stats.mean(zs)
```

```python
cen = pd.read_csv('/Users/guill/Desktop/Practica2/5cen.csv')
die = pd.read_csv('/Users/guill/Desktop/Practica2/10peso.csv')
alc = pd.read_csv('/Users/guill/Desktop/Practica2/Alcancia.csv')
gra = pd.read_csv('/Users/guill/Desktop/Practica2/Grapa.csv')
alu = pd.read_csv('/Users/guill/Desktop/Practica2/Aluminio.csv')
```

```python
a = np.zeros(6)
a[0] = np.min(zs)
a[1] = np.min(zcen)
a[2] = np.min(zdie)
a[3] = np.min(zgra)
a[4] = np.min(zalc)
a[5] = np.min(zalu)
x = np.arange(6)
```

```python
fig, ax = plt.subplots(figsize=(8, 5), constrained_layout=True)
plt.plot(h.iloc[:,0], zs, color= 'yellow', label='Marco de referencia')
plt.plot(alu.iloc[:,0], zalu, color ='blue', label = 'Paramagnético')
plt.plot(cen.iloc[:,0], zcen, color = 'orange', label= 'Ferromagnético')
plt.plot(die.iloc[:,0], zdie, color = '#40E0D0', label = 'Diamagnético')
plt.plot(gra.iloc[:,0], zgra, color = 'orange')
plt.plot(alc.iloc[:,0], zalc, color = 'orange')
plt.title('Campo magnético en función del tiempo para los distintos tipos de materiales magnéticos')
plt.xlabel('Tiempo (s)')
plt.ylabel('Campo magnético ($\mu T$)')
plt.legend()
plt.grid()

plt.savefig('Todos')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 5), constrained_layout=True)
plt.scatter(x[0], a[0], color= 'green', label='Marco de referencia')
plt.scatter(x[1], a[1], color = 'red',  label= 'Ferromagnético')
plt.scatter(x[2], a[2], color = '#40E0D0', label = 'Diamagnético')
plt.scatter(x[3], a[3], color = 'red')
plt.scatter(x[4], a[4], color = 'red')
plt.scatter(x[5], a[5], color ='blue', label = 'Paramagnético')
plt.title('Campo magnético en función del tiempo para los distintos tipos de materiales magnéticos')
plt.xlabel('Medición')
plt.ylabel('Campo magnético ($\mu T$)')
plt.legend()
plt.grid()

plt.savefig('Puntos')
plt.show()
```
