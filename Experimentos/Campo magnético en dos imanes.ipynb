```python
for m in range(5):    
    d = []
    for j in range(2,21,2):
        d.append(pd.read_csv('/Users/guill/Desktop/Proyecto/'+str(j)+'0'+str(m)+'.csv'))
        
    d2 = d[0]
    d4 = d[1]
    d6 = d[2]
    d8 = d[3]
    d10 = d[4]
    d12 = d[5]
    d14 = d[6]
    d16 = d[7]
    d18 = d[8]
    d20 = d[9]
    
    x2 = d2.iloc[:,1]
    x4 = d4.iloc[:,1]
    x6 = d6.iloc[:,1]
    x8 = d8.iloc[:,1]
    x10 = d10.iloc[:,1]
    x12 = d12.iloc[:,1]
    x14 = d14.iloc[:,1]
    x16 = d16.iloc[:,1]
    x18 = d18.iloc[:,1]
    x20 = d20.iloc[:,1]
    
    B = np.zeros(10)
    B[0] = stats.mean(x2)
    B[1] = stats.mean(x4)
    B[2] = stats.mean(x6)
    B[3] = stats.mean(x8)
    B[4] = stats.mean(x10)
    B[5] = stats.mean(x12)
    B[6] = stats.mean(x14)
    B[7] = stats.mean(x16)
    B[8] = stats.mean(x18)
    B[9] = stats.mean(x20)
    B = np.abs(B)

    dis = []
    for k in range(2,21,2):
        dis.append(k)
    dis = np.array(dis)
    dis = dis/100
    
    yer = np.zeros(10)
    yer[0] = stats.pstdev(x2)
    yer[1] = stats.pstdev(x4)
    yer[2] = stats.pstdev(x6)
    yer[3] = stats.pstdev(x8)
    yer[4] = stats.pstdev(x10)
    yer[5] = stats.pstdev(x12)
    yer[6] = stats.pstdev(x14)
    yer[7] = stats.pstdev(x16)
    yer[8] = stats.pstdev(x18)
    yer[9] = stats.pstdev(x20)
    
    Bl = np.log(B)
    disl = np.log(dis) 
    
    x = disl
    y = Bl

    n = len(y)

    g = int(1)

    sx = np.zeros(2*g)
    sxy = np.zeros(g+1)

    for j in range((2*g)):
        for i in range(n):
            sx[j] = x[i]**(j+1) + sx[j]

    for j in range((g+1)):
        for i in range(n):
            sxy[j] = x[i]**(j)*y[i]+sxy[j]

    A = np.zeros([g+1,g+1])
    X = np.zeros(g+1)
    B = np.zeros([g+1,1])

    for i in range(g+1):
        for j in range(g+1):
            if (i+j)!=0:
                A[i][j] = sx[i+j-1]
            if i==0 and j==0:
                A[i][j] = n

    for i in range(g+1):
        B[i] = sxy[i]    

    MAT = np.concatenate((A,B), axis=1)

    for j in range((g+1)-1):
        for i in range(j+1,(g+1)):
            MAT[i][:] = MAT[i][:] + MAT[j][:]*(-MAT[i][j]/MAT[j][j])

    for i in range((g+1)-1,-1,-1):
        suma = 0

        if i<n:
            for j in range((g+1)-1,-1,-1):
                suma = MAT[i][j]*X[j]+suma

        X[i] = (-suma + MAT[i][g+1])/MAT[i][i]

    yp = 0
    e = 0
    s = 0

    for i in range(n):
        yp = y[i]+yp

    yp = yp/n

    for i in range(n):
        sume = 0
        for j in range(g,-1,-1):
            sume = -X[j]*x[i]**(j) + sume
        e = (y[i] + sume)**2 + e
        s = (y[i]- yp)**2 + s

    #print("Los datos calculados fueron los siguientes:")
    #print(arch.T)
    r = math.sqrt((s-e)/s)
    
    name = ['Un sólo iman', 'Dos imanes juntos', 'Dos imanes. Separación 2 mm', 'Dos imanes. Separación 4 mm', 'Dos imanes. Separación 6 mm']
    print('Para '+str(name[m]))
    print("La recta calulada fue la siguiente y =")
    for i in range(g,-1,-1):
        print(str(X[i]) + "x^" + str(i))

    print('A es igual a: ' + str(np.exp(X[0])))
    print('n es igual a: ' +str(X[1]))
    #print("r calculada fue de: " + str(r))
    #print("E calculada fue de: " + str(e))   

    imp = np.zeros(n)

    for i in range(g,-1,-1):
        imp = X[i]*x**i + imp

    
    colores = ['#d92027', '#303960', '#ffd31d', '#79d70f', '#f37121']

    if m == 0:
        fig, ax = plt.subplots(figsize=(8, 5), constrained_layout=True)
        plt.title('Ajuste lineal de los datos obtenidos')
        plt.xlabel('Posición (m)')
        plt.ylabel('Campo magnético ($\mu T$)')
    
    plt.plot(x, imp, color=colores[m], label='Ajuste ln B = ln A + nln x '+str(name[m]))
    plt.scatter(x,y, color='black',s=10)

    if m == 4:
        plt.legend()
        plt.grid()
        plt.savefig('ajuste')
        plt.show()
        
```
