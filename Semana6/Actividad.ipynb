```python
path = '/Users/guill/Desktop/AC/da2020-seguragg2018/datos/'
data = np.load(path+'fit_exercise.npy')
```

```python
plt.errorbar(data[0], data[1], yerr=data[2])
```

```python
def f(x, m, b):
    return m*x+b

def chi2(m, b):
    return np.sum(((data[1] - f(data[0], m, b))**2)/data[2]**2)
```

```python
m = np.linspace(0,-2, 10)
b = np.linspace(2, 8, 10)
```

```python
#Malla 
n = np.zeros((10,10))
for i in range(10):
    for j in range(10):
        n[i][j] = chi2(m[i], b[j])
        
print("Chi minima = ",np.amin(n))
pos = np.where(n == np.amin(n))
print("m = ", m[pos[0][0]], "b = ", b[pos[1][0]])
```
