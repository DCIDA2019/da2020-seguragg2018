# Ajustes de datos

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
import requests
```

Existen métodos de ajustes de datos que solo consideran la ubicación de los puntos pero no sus posibles errores. El tipo de ajuste que vamos a implementar a continuación considera todo esto. Creo que se llama: método de $\chi^2$

```python
path = '/Users/guill/Desktop/AC/da2020-seguragg2018/datos/'
data = np.load(path+'fit_exercise.npy')
```

```python
plt.errorbar(data[0], data[1], yerr=data[2])
plt.grid()
```

```python
def f(x, m, b):
    return m*x+b

def chi2(m, b):
    return np.sum(((data[1] - f(data[0], m, b))**2)/data[2])
```

```python
def chi_s(theta):
    aux = 0
    for i in range(len(data[0])):
        aux = aux + (data[1][i]-(theta[0]*data[0][i]+theta[1]))**2/data[2][i]**2
    return aux
```

## Chi cuadrada para una malla uniforme

```python
n = 1000
m = np.linspace(-2,0, n)
b = np.linspace(2, 8, n)
```

```python
#Generamos una malla espaciada uniformemente
chi = np.zeros((n,n))
for i in range(n):
    for j in range(n):
        chi[i][j] = chi2(m[i], b[j])
        
print("Chi minima = ",np.amin(chi))
pos = np.where(chi == np.amin(chi))
print("m = ", m[pos[0][0]], "b = ", b[pos[1][0]])
m_min = m[pos[0][0]]
b_min = b[pos[1][0]]
```

```python
plt.errorbar(data[0], data[1], yerr=data[2])
x = np.arange(10)
plt.plot(m_min*x+b_min)

plt.title('Ajuste para una malla uniforme')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.grid()
```

## Chi cuadrada para puntos aleatorios

```python
#Generamos un monton de puntos aleatorios
n = 100000
p = np.random.random((n, 2))
p[:, 0] = p[:, 0]*(2)-2
p[:, 1] = p[:, 1]*(6)+2
p
```

```python
chial = np.zeros(n)
for i in range(n):
    chial[i] = chi_s(p[i])

print("Chi minima =", min(chial))
pos = np.where(chial==min(chial))
ma = p[pos[0][0]][0]
ba = p[pos[0][0]][1]
print("m = ", ma, "b =", ba)
```

```python
plt.errorbar(data[0], data[1], yerr=data[2])
x = np.arange(10)
plt.plot(ma*x+ba, color = 'r')

plt.title('Ajuste de puntos aleatorios')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.grid()
```

## Chi cuadrada utilizando la función 'optimize'

```python
x = [0, 7]
res = minimize(chi_s, x)

mo = res.x[0]
bo = res.x[1]

print("m = ", mo, "b =", bo)
```

```python
plt.errorbar(data[0], data[1], yerr=data[2])
x = np.arange(10)
plt.plot(mo*x+bo, color = 'black')

plt.title('Ajuste para optimize')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.grid()
```

Vamos a comparar los tres resultados

```python
plt.errorbar(data[0], data[1], yerr=data[2])
x = np.arange(10)

plt.plot(mo*x+bo, color = 'black', label='Optimize')
plt.plot(ma*x+ba, color = 'r', label ='Datos aleatorios')
plt.plot(m_min*x+b_min, label='Malla uniforme')

plt.title('Comparación de los ajustes')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.legend()
plt.grid()
```
