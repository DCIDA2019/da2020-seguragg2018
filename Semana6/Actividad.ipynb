```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
path = '/Users/guill/Desktop/AC/da2020-seguragg2018/datos/'
data = np.load(path+'fit_exercise.npy')
```

```python
plt.errorbar(data[0], data[1], yerr=data[2])
plt.grid()
```

```python
theta = [np.zeros(len(data[0])),np.zeros(len(data[0])),np.zeros(len(data[0]))]
```

```python
def f(x, m, b):
    return m*x+b

def chi2(m, b):
    return np.sum(((data[1] - f(data[0], m, b))**2)/data[2])
```

```python
n = 100
m = np.linspace(0,-2, n)
b = np.linspace(2, 8, n)
theta[0] = data[0]
theta[1] = m
theta[2] = b
```

```python
#Malla
chi = np.zeros((n,n))
for i in range(n):
    for j in range(n):
        chi[i][j] = chi2(m[i], b[j])
        
print("Chi minima = ",np.amin(chi))
pos = np.where(chi == np.amin(chi))
print("m = ", m[pos[0][0]], "b = ", b[pos[1][0]])
m_min = m[pos[0][0]]
b_min = b[pos[1][0]]
```

```python
plt.errorbar(data[0], data[1], yerr=data[2])
x = np.arange(10)
plt.plot(m_min*x+b_min)
plt.grid()
```
