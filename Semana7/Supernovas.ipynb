$\mu = m_B^* -(M_B-a \times X_1 \times \beta \times C)$

```python
func = np.zeros(740)
def mus(a, b, mb):
    for i in range(740):
        func[i] = data.iloc[i,4]-(mb-a*data.iloc[i,6]*b*data.iloc[i,8])
    return func
```

```python
a =  0.141
b = 3.101
mb = -19.05
```

```python
plt.scatter(data['zcmb'],(mus(a, b, mb)))
plt.xscale('log')
plt.xlim(0.001,10)

plt.title('Mu en función de z')
plt.xlabel('z')
plt.ylabel('$\mu$')

plt.grid()
```
