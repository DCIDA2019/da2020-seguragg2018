```python
def f(x):
    print("La funcion se llama f")
    return x**2 + 2*x - 5
```

```python
b = np.arange(-5,5,0.01)
```

```python
plt.plot(b,f(b), color='r')
plt.xlabel("X")
plt.ylabel("f(X)")
plt.title("Gráfica de una función f(x)")
plt.grid()
plt.legend("Fución")
plt.savefig("Figura.jpg")
```
