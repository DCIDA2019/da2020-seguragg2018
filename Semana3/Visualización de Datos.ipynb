# Visualización con Matplotlib y Pandas

```python
import numpy as np
import pandas as pd
import requests
import os
import matplotlib.pyplot as plt
```

Utilizando las herramientas adquiridas con la libreria 'Pandas' es necesario también conocer como podemos representar los datos de las tablas de manera gráfica. Podemos hacer uso de la libreria 'Matplotlib' que nos permite crear una gran gama de representaciones gráficas. 

```python
tabulador = pd.read_excel('../datos/tabulador_pe2019.xls', header = 9)
```

```python
tabulador
```

Podemos visualizar los datos anteriores en base a diferentes representaciones gráficas para obtener distintas informaciones. Podemos graficar el sueldo base para darnos una idea de cual es la distribución en los sueldos de los funcionarios. Hay registros cuyo sueldo es de 60,000 mientras que hay otros registros que no superan los 10,000, es curioso ver que estos últimos son casi la mitad, aunque para ver esto de manera mas clara necesitamos recurrir a otro tipo de gráfico.

```python
plt.plot(tabulador['SUELDO BASE'], marker='o',color='black')
plt.xlabel('Número de registro')
plt.ylabel('Sueldo ($)')
plt.title('Sueldo base en función del número de registro')
plt.grid()
```

Otro tipo de dato, muy interesante por cierto, que podemos obtener de la columna 'SUELDO BASE' es la distribución de los distintos sueldos en la cantidad de funcionarios. 

```python
plt.hist(tabulador['SUELDO BASE'], color='red')
plt.xlabel('Sueldo ($)')
plt.ylabel('Cantidad de funcionarios')
plt.title('Cantidad de funcionarios en función del sueldo base')
plt.grid()
```

Podemos incluso hacer un gráfico de pastel para ver de una manera diferente la distribución del salario base.

```python
plt.pie(tabulador['SUELDO BASE'], labels=tabulador['PUESTO'])
plt.title("Sueldo base de los funcionarios públicos")
plt.show()
```

```python
plt.pie(tabulador['SUELDO MENSUAL NETO'], labels=tabulador['PUESTO'])
plt.title("Sueldo base de los funcionarios públicos")
plt.show()
```

Es evidente que no es una buena manera de visualizar estos datos pues son demasiados, aunque en otros casos podría ser muy útil. :D

La funcion 'describe' nos da una descripción estadística de la tabla, lo cual es bastante útil.

```python
tabulador.describe()
```

Por ejemplo, una cosa que flipa es comparar los historgramas. O inclusive también podemos hacerlo con las gráficas

```python
kwargs = dict(histtype='stepfilled', alpha=0.3)
plt.hist(tabulador['SUELDO BASE'], **kwargs, label = 'Sueldo base')
plt.hist(tabulador['SUELDO MENSUAL NETO'], **kwargs, label = 'Sueldo mensual neto')
plt.hist(tabulador['SUELDO MENSUAL BRUTO'], **kwargs, label = 'Sueldo mensual bruto')
plt.xlabel('Sueldo ($)')
plt.ylabel('Cantidad de funcionarios')
plt.title('Cantidad de funcionarios en función del sueldo')
plt.legend()
plt.grid()
```

```python
plt.plot(tabulador['SUELDO BASE'], color='black', label = 'Sueldo base')
plt.plot(tabulador['SUELDO MENSUAL NETO'], color='blue', label = 'Sueldo mensual neto')
plt.plot(tabulador['SUELDO MENSUAL BRUTO'], color='red', label = 'Sueldo mensual bruto')
plt.xlabel('Número de registro')
plt.ylabel('Sueldo ($)')
plt.title('Sueldo en función del número de registro')
plt.legend()
plt.grid()
```

Podemos hacer un análisis similar para los siguientes datos, en los cuales los únicos parámetros en los que podemos indagar son los que estan en las últimas tres columnas.

```python
file = 'salarios_minimos.csv'
if not os.path.isfile(file):
    url = 'https://idconline.mx/laboral/salarios-minimos/salarios-minimos-1986-2007'
    html = requests.get(url).content
    df_list=pd.read_html(html,header=0)
    df = df_list[-1]
    df.to_csv('salarios_minimos.csv')
else:
    df = pd.read_csv(file).drop(columns='Unnamed: 0')
    
df
```

Vamos a hacer un análisis de los salarios mínimos con el paso del tiempo de la columna "A". Se divide el primer periodo entre 1000 por la devaluación que sufrió el peso en aquel entonces :(. Primero, calculamos el promedio de los salarios mímimos en los periodos de tiempo.

```python
Va = df.iloc[:,4]
V = np.mean(Va[36:45])/1000

Ma = df.iloc[:,4]
M = np.mean(Ma[24:36])

Ca = df.iloc[:,4]
C = np.mean(Ca[11:24])

Na = df.iloc[:,4]
N = np.mean(Na[0:11])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

nombres = ["1988 - 1992",'1993 - 2000','2001 - 2012','2013 - 2019']
datos = [V,M,C,N]
xx = range(len(datos))

ax.bar(xx, datos, width=0.5, align='center')
ax.set_xticks(xx)
ax.set_xticklabels(nombres)

plt.xlabel('Periodo de tiempo')
plt.ylabel('Salario mínimo ($)')
plt.title('Promedios del salario mínimo a lo largo del tiempo')
plt.grid()
plt.show()
```
