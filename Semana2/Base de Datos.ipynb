# Resumenes mensuales de lluvia y temperaturas

El ejericio consiste en realizar una base de datos utilizando el material de la comisión nacional del agua empleando las herramientas que nos da la libreria pandas. Realizaremos una base de datos para la temperatura y otra para las lluvias.

```python
import pandas as pd
import numpy as np
```

## Temperaturas

Se utilizarán los datos de temperaturas de la conagua de enero de 2020


```python
T_max = pd.read_csv('T_Max.csv', encoding='latin-1')
T_med = pd.read_csv('T_Med.csv', encoding='latin-1')
T_min = pd.read_csv('T_Min.csv', encoding='latin-1')
```

```python
pd.merge(pd.merge(T_med, T_min),T_max)
```

```python
print(len(T_max))
print(len(T_med))
print(len(T_min))
```

Nótese que el dataFrame de la temperatura máxima tiene menos elemenos que los dataframes de la temperatura media y mínima, que tienen la misma cantidad de datos, por lo que cuando hacemos merge encontramos la intersección entre media, mínima y máxima, y el dataFrame final tiene menos elemenos que si hacemos merge entre minima y media, que es ideal, pues tiene el mismo número de elementos que media y mínima individualmente.

```python
pd.merge(T_med, T_min)
```

## Lluvias

Para la base de datos de las lluvias se utilizarán los datos de la conagua de los años 2018, 2019 y 2020 del mes de enero

```python
l_2018 = pd.read_csv('Lluvias2018.csv', encoding='latin-1')
l_2019 = pd.read_csv('Lluvias2019.csv', encoding='latin-1')
l_2020 = pd.read_csv('Lluvias2020.csv', encoding='latin-1')
```

La última columna del DataFrame marca el volumen de lluvias del mes de enero; en cada uno de los datos de 2018, 2019 y 2020 esta columna tiene el mismo nombre, por lo que cuando hacemos un merge, no tenemos el resultado esperado. Además, en el caso del año 2018, en lugar de que la columna de la entidad federativa diga 'EDO' la columna dice 'ESTADO'.
Tenemos que renombrar las columnas.

```python
l_2018.rename(columns={'ENE':'ENE2018', 'ESTADO':'EDO'}, inplace=True)
l_2019.rename(columns={'ENE':'ENE2019'}, inplace=True)
l_2020.rename(columns={'ENE':'ENE2020'}, inplace=True)
```

Ahora, cuando hacemos merge, la base de datos nos mostrará los resultados de la última columna separados.

```python
pd.merge((pd.merge(l_2018, l_2019)),l_2020)
```
