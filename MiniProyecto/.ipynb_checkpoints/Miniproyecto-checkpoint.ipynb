# ¿Cómo morimos en México?

Para este proyecto se utilizarán datos del INEGI para las muertes en México. Los datos se pueden encontrar en: https://www.inegi.org.mx/temas/mortalidad/. En dicha base de datos se tiene información sobre el volumen de las defunciones registradas en el país, así como algunas características por edad y sexo de los fallecidos y las principales causas que originan los decesos.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
m = pd.read_excel('mortalidad_00.xlsx')
```

```python
m
```

Cuando nos ponemos a indagar en la base de datos nos damos cuenta que primero tiene información general de toda la república, y después tiene datos para cada estado y municipios respectivos, aunque si somos mas minuciosos, cada estado tiene información general de 12 indicadores: 
1. Defunciones generales
2. Defunciones generales hombres
3. Defunciones generales mujeres
4. Defunciones generales sexo no especificado
5. Defunciones de menores de un año
6. Defunciones de menores de un año hombres
7. Defunciones de menores de un año mujeres
8. Defunciones de menores de un año sexo no especificado
9. Suicidios registrados
10. Índice de sobremortalidad masculina
11. Porcentaje de muertes por suicidio con respecto al total de muertes violentas
12. Porcentaje de hijos fallecidos de las mujeres de 15 a 49 años

Los datos generales para la república mexicana tienen muchos mas indicadores, entre los cuales podemos destacar además de los anteriores, los siguientes:
1. Defunciones por homicidio. Hombre
2. Defunciones por homicidio. Mujer
3. Tasa de defunciones por homicidio por cada 100 mil habitantes, mujeres
4. Tasa de defunciones por homicidio por cada 100 mil habitantes, hombres
5. Defunciones por homicidio
6. Tasa de defunciones por homicidio por cada 100 mil habitantes


```python
muj = m.iloc[17,:]
muj = muj.drop(['cve_entidad', 'desc_entidad', 'cve_municipio', 'desc_municipio', 'id_indicador', 'indicador', 'unidad_medida'], axis=0)
```

```python
hom = m.iloc[13,:]
hom = hom.drop(['cve_entidad', 'desc_entidad', 'cve_municipio', 'desc_municipio', 'id_indicador', 'indicador', 'unidad_medida'], axis=0)
```

```python
fig, ax = plt.subplots(figsize=(8, 5), constrained_layout=True)

plt.plot(muj, label ='Mujeres')
plt.plot(hom, label ='Hombres')

plt.xlabel('Año')
plt.ylabel('Número de muertos')
plt.title('Homicidios totales en México')
plt.setp(ax.get_xticklabels(), rotation=90, ha="right")
plt.legend()
plt.grid()

plt.show()
```
