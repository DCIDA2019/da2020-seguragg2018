# ¿Cuántos morimos en México?

Para este proyecto se utilizarán datos del INEGI para las muertes en México. Los datos se pueden encontrar en: https://www.inegi.org.mx/temas/mortalidad/. En dicha base de datos se tiene información sobre el volumen de las defunciones registradas en el país, así como algunas características por edad y sexo de los fallecidos y las principales causas que originan los decesos.

## Preguntas

Podemos hacer preguntas que refieran a los datos generales de México y posteriormente tratar los estados para finalmente relacionar ambos. \\

```python
defg = m.iloc[0,:]
defg = defg.drop(['cve_entidad', 'desc_entidad', 'cve_municipio', 'desc_municipio', 'id_indicador', 'indicador', 'unidad_medida'], axis=0)
defgh = m.iloc[1,:]
defgh = defgh.drop(['cve_entidad', 'desc_entidad', 'cve_municipio', 'desc_municipio', 'id_indicador', 'indicador', 'unidad_medida'], axis=0)
defgm = m.iloc[2,:]
defgm = defgm.drop(['cve_entidad', 'desc_entidad', 'cve_municipio', 'desc_municipio', 'id_indicador', 'indicador', 'unidad_medida'], axis=0)
defgs = m.iloc[3,:]
defgs = defgs.drop(['cve_entidad', 'desc_entidad', 'cve_municipio', 'desc_municipio', 'id_indicador', 'indicador', 'unidad_medida'], axis=0)
```

```python
fig, ax = plt.subplots(figsize=(8, 5), constrained_layout=True)

plt.plot(defg, label ='Generales', color='black')
plt.plot(defgh, label ='Hombres', color='red')
plt.plot(defgm, label = 'Mujeres', color='green')
plt.plot(defgs, label = 'Sexo no especificado', color='blue')

plt.xlabel('Año')
plt.ylabel('Número de muertos')
plt.title('Defunciones generales en México')
plt.setp(ax.get_xticklabels(), rotation=90, ha="right")
plt.legend()
plt.grid()

plt.show()
```
