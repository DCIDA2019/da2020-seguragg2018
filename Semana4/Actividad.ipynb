```python
import numpy as np
import pandas as pd
import requests
import os
import matplotlib.pyplot as plt
```

```python
tabulador = pd.read_excel('../datos/tabulador_pe2019.xls', header = 9)
tabulador
```

```python
plt.plot(tabulador['SUELDO MENSUAL NETO'], tabulador['IMPTO MENSUAL'], color = 'red')
plt.xlabel('Sueldo de cada funcionario ($)')
plt.ylabel('Cantidad de impuestos ($)')
plt.title('Impuestos en función del sueldo mensual neto')
plt.grid()
plt.savefig("ImpuestosSueldoMN.jpg")
```

```python
plt.plot(tabulador['SUELDO BASE'], tabulador['IMPTO MENSUAL'])
plt.xlabel('Sueldo de cada funcionario ($)')
plt.ylabel('Cantidad de impuestos ($)')
plt.title('Impuestos en función del sueldo base')
plt.grid()
plt.savefig("ImpuestosSueldoB.jpg")
```

```python
plt.plot(tabulador['SUELDO MENSUAL BRUTO'], tabulador['IMPTO MENSUAL'], color='Black')
plt.xlabel('Sueldo de cada funcionario ($)')
plt.ylabel('Cantidad de impuestos ($)')
plt.title('Impuestos en función del sueldo mensual bruto')
plt.grid()
plt.savefig("ImpuestosSueldoSMB.jpg")
```

```python
plt.plot(tabulador['SUELDO BASE'], tabulador['GRATIFICACIÓN QUINCENAL'], color='Green')
plt.xlabel('Sueldo de cada funcionario ($)')
plt.ylabel('Gratificación Quincenal ($)')
plt.title('Gratificación quincenal en función del sueldo base')
plt.grid()
plt.savefig("GratificacionSB.jpg")
```

```python
kwargs = dict(histtype='stepfilled', alpha=0.3)
plt.hist(tabulador['SUELDO BASE'], **kwargs, label = 'Sueldo base')
plt.hist(tabulador['SUELDO MENSUAL NETO'], **kwargs, label = 'Sueldo mensual neto')
plt.hist(tabulador['SUELDO MENSUAL BRUTO'], **kwargs, label = 'Sueldo mensual bruto')
plt.xlabel('Sueldo ($)')
plt.ylabel('Cantidad de funcionarios')
plt.title('Cantidad de funcionarios en función del sueldo')
plt.legend()
plt.grid()
plt.savefig("CantSueldo.jpg")
```

```python
plt.plot(tabulador['SUELDO BASE'], tabulador['SUELDO MENSUAL NETO'], color='#ffcc00')
plt.xlabel('Sueldo base ($)')
plt.ylabel('Sueldo mensual neto ($)')
plt.title('Sueldo mensual neto en función del sueldo base')
plt.grid()
plt.savefig("BaseMSN.jpg")
```

```python
plt.plot(tabulador['SUELDO BASE'], tabulador['SUELDO MENSUAL NETO'], color='#ffcc00')
plt.xlabel('Sueldo base ($)')
plt.ylabel('Sueldo mensual neto ($)')
plt.title('Sueldo mensual neto en función del sueldo base')
plt.grid()
plt.xlim(0,20000)
plt.ylim(0,40000)
plt.savefig("BaseMSN1.jpg")
```
